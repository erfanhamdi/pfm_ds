```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import glob
plt.style.use('/projectnb/lejlab2/erfan/Recode/documentation/theme.mplstyle')
import matplotlib.font_manager as font_manager
path = '/projectnb/lejlab2/erfan/Recode/documentation/Times_New_Roman.ttf'
prop = font_manager.FontProperties(fname=path)
mpl.rcParams['font.family'] = prop.get_name()
```

```python
plt.figure(figsize=(10, 8))
force_disps = glob.glob("/projectnb/lejlab2/erfan/dataset_phase2/Generalized/results/val/quad_1*_VV_256/*_bot.txt")
force_disps = [
"/projectnb/lejlab2/erfan/dataset_phase2/Generalized/results/val/quad_1_hyb_miehe_VV_256/force_bot.txt",
"/projectnb/lejlab2/erfan/dataset_phase2/Generalized/results/val/quad_1_hyb_amor_VV_256/force_bot.txt",
"/projectnb/lejlab2/erfan/dataset_phase2/Generalized/results/val/quad_1_hyb_star_VV_256/force_bot.txt",
"/projectnb/lejlab2/erfan/pfm_ds/verification/results/2d_shear/force_bot_y.txt",
"/projectnb/lejlab2/erfan/pfm_ds/verification/results/2d_shear_amor/force_bot_x.txt",
]
verif = np.loadtxt("/projectnb/lejlab2/erfan/Recode/documentation/verification/shear_ambati_hybrid_2015.csv", delimiter=',')
verif_mask = np.where(verif[:, 0] < 0.02)
plt.plot(verif[:28, 0], verif[:28, 1], label="Ambati et al. 2015", linestyle="-")
print(force_disps)
Label_list = ['Miehe', 'Amor', 'Star', 'miehehehe', 'amorres']
for idx, fd in enumerate(force_disps):
    print(fd)
    force_disp = np.loadtxt(fd)
    # label = fd.split("/")[-2].split("_")[3]
    plt.plot(force_disp[:, 1], -force_disp[:, 0]*1e-3, label=Label_list[idx])
plt.legend()
plt.xlabel("Displacement [mm]")
plt.ylabel("Force [kN]")
plt.tight_layout()
```
